```python
#store the resume in variable
resume = docx2txt.process("python-developer-resume-demo (2).docx")
#print resume
print(resume)
```

```python
#store job description  into a veriable
job_description = docx2txt.process("Software Test Engineer Job Description.docx")
#print job description
print(job_description)
```

```python
#list of text
text = [resume, job_description]
```

```python
#import library to count the number of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
#import library for similar function
from sklearn.metrics.pairwise import cosine_similarity

#print hte similarity scores
print("\nSimilarity Scores:")
print(cosine_similarity(count_matrix))
```

similarity function shows that these two files have 0.529% similar

```python
#get the match percentage
matchpercentage = cosine_similarity(count_matrix)[0][1]*100
#converting value into 2 decimal
matchpercentage = round(matchpercentage, 2)
print("your resume match about "+ str(matchpercentage)+ "% of the job description")
```
